## Tools

Models can request to call these tools that perform tasks such as fetching data from database, searching the web, or running code. Tools are made up of:
    
    1. A schema, including the name of tool, description, and\or argument definition (often JSON schema)
    2. A function or coroutine to execute

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3.6-27b")    
response = model.invoke("Why do parrots talk?")
response
```

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """Get the current weather in a given location."""
    # In a real implementation, you would call a weather API here.
    return f"The current weather in {location} is sunny with a temperature of 25°C."

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What is the weather in New York?")
print(response)
for tool_call in response.tool_calls:
    print(f"Tool: {tool_call['name']}")
    print(f"Tool: {tool_call['args']}")
```
