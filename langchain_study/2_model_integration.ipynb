## Streaming

nearly all llms allow us to show output as it is being generated, which can be displayed to improve user experience


```python
## the output is generated and displayed simultaneously as the model generates it. This is useful for long outputs where you want to see the response in real-time.

for chunk in model.stream("Write a 300 word paragraph on football"):
    print(chunk.content, end="", flush=True)
```

## Batching

batching is collection of independent requests to models can improve performance and reduce cost, processing done in parallel


```python
model.batch(
    ["Why do parrots talk?",
      "Why is the sky blue?",
     "What is the capital of France?"
    ],
    config = {
        'max_concurrency': 5,
    }
)
```
