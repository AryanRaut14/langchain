### Model integration of OPENAI, GROQ and GOOGLE

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gpt-4.1", temperature=0.7)
model
```

# gemini model integration

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-3.5-flash-lite")
response = model.invoke("What is the capital of India?")
response.content
```

## groq model integration

```python
import os 
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY") 

model = init_chat_model("groq:qwen/qwen3.6-27b")
response = model.invoke("Why is sun white?")
response
```

```python
import os 
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3.6-27b")
response = model.invoke("Why do parrots talk?")
response
```

There are 2 methods for model integration

1. using init_chat_model
2. using specific to their own models
